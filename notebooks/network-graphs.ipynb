```python
import networkx as nx
import pandas as pd
```

```python
G = nx.Graph()
```

```python
df = pd.read_csv('../data-display/misassigned-connections.csv')
df.head()
```

```python
G.add_nodes_from(df['dept_from'].unique().tolist())
```

```python
edges = dict()
for x in df.values:
    e = sorted([x[0], x[1]])
    e = '<=>'.join(e)
    if e in edges:
        tickets = edges[e] + x[2]
        edges[e] = tickets
    else:
        edges[e] = x[2]   
```

```python
for key, value in edges.items():
    es = key.split('<=>')
    G.add_edge(es[0], es[1], weight=value)
```
